# Vehículos, autos y bicicletas:
*	Color: Default: Blanco
*	Velocidad promedio
*	Kilómetros recorridos: Default: 0 km
*	Kilómetros máximos por recorrer
*	Contador de ciudades visitadas
*	Estado: Funcionando o En Panne: Default: Funcionando

Autos:
 > -	Km recorridos: máx. de 1500km
 > -	Velocidad promedio de 80 km/hr
 
Bicicletas:
 > -	Km máx. por recorrer 1000km
 > -	Velocidad promedio: 30 km/hr
 
*	Métodos:
 > - Eficiencia: ciudades visitadas / km recorridos
 > - Ultima ciudad visitada
 > - Avanzar, retorna tiempo en recorrer distancia y se puede generar un panne según distancia máxima
 > - Print, el tipo del vehículo, su color actual del vehículo, los kilómetros recorridos y el estado actual.


```python
class Ciudad:
    
    def __init__(self, name, color, painted_vehicles=0):
        self.name = name
        self.color = color
        self.painted_vehicles = painted_vehicles

    def pintar(self, vehicle):
        self.painted_vehicles += 1
        vehicle.visited_cities += 1
        vehicle.last_visited_city = self.color

    def __str__(self):
        return "{}: color {}, {} vehículos pintados.".format(self.name, self.color, self.painted_vehicles)


class Vehicle:

    def __init__(self, avg_speed, km_max, color='blanco', km=0, visited_cities=0, state="Funcionando"):
        self.avg_speed = avg_speed
        self.km_max = km_max
        self.color = color
        self.km = km
        self.visited_cities = visited_cities
        self.state = state

    def efficiency(self):
        return self.visited_cities/self.km
    
    def last_visited_city(self):
        return self.color

    def recorrer(self, km):
        if self.km + km <= self.km_max:
            self.km += km
        else:
            self.km = self.km_max
            self.state = "En Panne"

    def str(self, type):
        return "{}: color actual {}, he recorrido {} kilómetros, estado es {}.".format(type, self.color, self.km, self.state)

class Auto(Vehicle):

    def __init__(self, *args):
        super().__init__(80, 1500, *args)

    def __str__(self):
        return super().str(self.__class__.__name__)

class Bicicleta(Vehicle):

    def __init__(self, *args):
        super().__init__(30, 1000, *args)

    def __str__(self):
        return super().str(self.__class__.__name__)
```
